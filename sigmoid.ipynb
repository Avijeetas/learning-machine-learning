```python
# http://squall0032.tumblr.com/post/77300791096/plotting-a-sigmoid-function-using
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5., 5., 0.5)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```
